```python
import requests
from bs4 import BeautifulSoup

site = requests.get('https://www.scrapethissite.com/pages/simple/').text

html_parsed = BeautifulSoup(site, 'html.parser')

```

```python
html_parsed


html_parsed.find(class_ = 'country-name').text.strip()
html_parsed.find(class_ = 'country-capital').text.strip()
```

```python
lista_paises = html_parsed.find_all(class_ = 'country-name')

lista_capitales = html_parsed.find_all(class_ = 'country-capital')

lista_final = []

for pais, capital in zip(lista_paises, lista_capitales):
    dici = { "pais": pais.text.strip(), 
            "capital": capital.text.strip()}
    
    lista_final.append(dici)
```
